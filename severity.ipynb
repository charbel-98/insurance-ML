```python
import pandas as pd
df = pd.read_csv('severity_dataset.csv')
# Drop rows with missing data in column: 'DATE OF \r\nACCIDENT'
df = df.dropna(subset=['DATE OF \r\nACCIDENT'])
# Drop rows with missing data in column: 'CLAIM_TYPE'
df = df.dropna(subset=['CLAIM_TYPE'])
df
```

```python
grouped = df.groupby(['POLICY_NO','DATE OF \r\nACCIDENT'])['DATE OF \r\nACCIDENT'].count()
print(grouped.describe())
```

```python
df.drop(columns=['POLICY_NO','DATE OF \r\nACCIDENT'], inplace=True)
```

```python
df
```

```python
df = df[df['SUM INSURED'] != 0]
df
```

```python
# apply random forest regressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np

# get all non numerical columns
non_numerical_cols = df.select_dtypes(exclude=[np.number]).columns.tolist()
# label encode all non numerical columns
le = LabelEncoder()
for col in non_numerical_cols:
    # use .loc
    df.loc[:, col] = le.fit_transform(df[col])
    # df[col] = le.fit_transform(df[col])


X = df.drop(columns=['SUM INSURED'])
y = df['SUM INSURED']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python

y_pred = model.predict(X_test)
print('R2: ', model.score(X_test, y_test))
print('MSE: ', mean_squared_error(y_test, y_pred))
print('RMSE: ', np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Create an array of indices for the x-axis
indices = np.arange(len(y_test))

# Plot actual values
plt.scatter(indices, y_test, label='Actual', color='blue', alpha=0.7)

# Plot predicted values
plt.scatter(indices, y_pred, label='Predicted', color='red', alpha=0.7)

# Add labels and title
plt.title('Actual vs. Predicted Values')
plt.xlabel('Data Point Index')
plt.ylabel('Frequency')
plt.legend()

plt.show()

```
